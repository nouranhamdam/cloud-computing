```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/books.csv')
```

```python
# Data Cleaning
df.drop_duplicates(inplace=True)
df.fillna(0, inplace=True)
```

```python
# Data Preprocessing
harry_potter_df = df[df['title'].str.contains('Harry Potter', case=False)]
```

```python
# Analysis
most_selling_books = harry_potter_df.sort_values(by='ratings_count', ascending=False).head()
average_rating = harry_potter_df['average_rating'].mean()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(most_selling_books['title'], most_selling_books['average_rating'], color='skyblue')
plt.title('Average Rating of Most Selling Harry Potter Books')
plt.xlabel('Book Title')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(most_selling_books['ratings_count'], labels=most_selling_books['title'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings for Most Selling Harry Potter Books')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
